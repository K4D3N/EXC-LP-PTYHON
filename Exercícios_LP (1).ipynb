```python
# 4 - Faça um Programa que peça as 4 notas bimestrais e mostre a média.
notas = [7, 5, 8, 10]
soma = 0
for notas in notas:
  soma += notas
media = soma/(len(str(notas)))
print('A média é:', media)
```

```python
# 9 - Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius. C = 5 * ((F-32) / 9).
print('Informe a Temperatura em Fahrenheit')

temperatura_fht = float(input())
temperatura_cls = 5 * ((temperatura_fht - 32)/9)
print('A temperatura em Celsius é:', temperatura_cls)
```

```python
# 10 - Faça um Programa que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.
print('Informe a Temperatura em Celsius')

temperatura_cls = float(input())
temperatura_fht = ((9*temperatura_cls)+160)/5
print('A temperatura em Fahrenheit é:', temperatura_fht)
```
